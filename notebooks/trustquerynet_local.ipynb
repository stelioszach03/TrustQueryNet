# TrustQueryNet Local Runner

This notebook is the thin local orchestrator. All core logic lives in the package so the same code can later run in Colab or on an A100 VM.

Use this notebook with a kernel that can already see the repository files.

- Local VS Code / local Python kernel: use this notebook.
- VS Code attached to a Colab kernel: use `notebooks/trustquerynet_colab.ipynb` instead, because the Colab runtime starts in `/content` and does not automatically have your local repo files.

```python
import os
import subprocess
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "pyproject.toml").exists():
    raise RuntimeError(
        f"This notebook expects the kernel to see the repo files. Current working directory is {PROJECT_ROOT}, "
        "which does not contain pyproject.toml. If you are attached to a Colab kernel, open notebooks/trustquerynet_colab.ipynb instead."
    )

os.chdir(PROJECT_ROOT)
subprocess.check_call([sys.executable, "-m", "pip", "install", "-e", f"{PROJECT_ROOT}[dev]"])
print(f"Using project root: {PROJECT_ROOT}")
```
